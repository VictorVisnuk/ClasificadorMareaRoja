## Lectura e intregración de datos meteorológicos

**Importación de librerias**

```python
import pandas as pd
```

**Lectura de ambos sets de datos (misma rango temporal distintas variables)**

```python
df_1 = pd.read_csv("POWER LARC 1.csv")
df_1.head()
```

```python
df_2 = pd.read_csv("POWER LARC 2.csv")
df_2.head()
```

**Unión de ambos sets de datos**

```python
df_unido = pd.merge(df_2, df_1, on=['YEAR', 'MO', 'DY'], how='inner')
df_unido.head()
```

**Tratamiento de las columnas de fechas para que sean compatibles a los datos de toxinas**

```python
df_unido['Fecha'] = pd.to_datetime(
    df_unido[['YEAR', 'MO', 'DY']].rename(columns={'YEAR': 'year', 'MO': 'month', 'DY': 'day'})
)
df_unido.head()
```

```python
df_unido= df_unido.drop(columns=['YEAR', 'MO', 'DY'])

# Reordenar columnas, poniendo 'Fecha' al principio
columnas = ['Fecha'] + [col for col in df_unido.columns if col != 'Fecha']
df_unido = df_unido[columnas]
df_unido.head()
```

```python
print("Calculando features de medias móviles...")

# Asegurarnos de que esté ordenado por fecha (crucial para .shift() y .rolling())
df_unido = df_unido.sort_values(by='Fecha')

# --- Definir variables y ventanas ---
# Usaremos 7 días (1 sem), 14 días (2 sem) y 30 días (1 mes)
ventanas = [7, 14, 30]

# Variables para medias (.mean())
vars_mean = [
    'T2M', 'T2M_MAX', 'T2M_MIN', # Temperatura
    'ALLSKY_SFC_PAR_TOT', 'ALLSKY_KT', # Radiación
    'WS2M' # Viento
]

# Variables para sumas (.sum())
vars_sum = [
    'PRECTOTCORR' # Precipitación
]

# --- Evitar Fuga de Datos (Data Leakage) ---
# Usamos .shift(1) para que los datos de un día (ej. 'Lunes')
# solo usen información *hasta* el día anterior ('Domingo').
# Esto es fundamental para que el modelo pueda predecir.

for v in ventanas:
    # Calcular Medias
    for var in vars_mean:
        col_name = f'{var}_MA_{v}d'
        df_unido[col_name] = df_unido[var].shift(1).rolling(window=v).mean()
        
    # Calcular Sumas
    for var in vars_sum:
        col_name = f'{var}_SUM_{v}d'
        df_unido[col_name] = df_unido[var].shift(1).rolling(window=v).sum()

# --- Manejo de NaNs ---
# Las primeras 'N' filas de cada nueva columna tendrán NaN (ej. los 1ros 6 días de MA_7d).
# Los rellenamos con el primer valor válido (bfill - backward fill).
# Esto solo afecta el inicio del dataset y es una práctica común.
print("Rellenando NaNs iniciales...")
df_unido = df_unido.bfill()

print("¡Features de medias móviles creadas!")
df_unido.head()
```

```python
df_unido = df_unido[df_unido['Fecha'] >= '2021-01-01']
df_unido.head()
```

**Exportación**

```python
df_unido.to_csv('POWER_LARC_unido.csv', index=False)
```
